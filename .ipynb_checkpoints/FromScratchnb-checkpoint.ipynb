```python
model = Sequential()

#1st Convolution:

def conv2d(input_image, kernel, stride=(1, 1), padding=(1, 1)):
    """
    Perform 2D convolution operation.

    Args:
        input_image (numpy.ndarray): Input image.
        kernel (numpy.ndarray): Convolution kernel.
        stride (tuple): Stride of the convolution operation in (vertical, horizontal) direction.
        padding (tuple): Padding to be applied in (vertical, horizontal) direction.

    Returns:
        numpy.ndarray: Convolved output.
    """
    # Get input image dimensions
    input_height, input_width = input_image.shape
    # Get kernel dimensions
    kernel_height, kernel_width = kernel.shape
    # Calculate output dimensions
    output_height = (input_height + 2 * padding[0] - kernel_height) // stride[0] + 1
    output_width = (input_width + 2 * padding[1] - kernel_width) // stride[1] + 1
    # Initialize output
    output = np.zeros((output_height, output_width))
    # Apply padding to input image
    padded_image = np.pad(input_image, ((padding[0], padding[0]), (padding[1], padding[1])), mode='constant')
    # Perform convolution
    for y in range(0, output_height):
        for x in range(0, output_width):
            # Apply stride
            y_start = y * stride[0]
            y_end = y_start + kernel_height
            x_start = x * stride[1]
            x_end = x_start + kernel_width
            # Extract region of interest from the padded image
            roi = padded_image[y_start:y_end, x_start:x_end]
            # Perform element-wise multiplication and sum
            output[y, x] = np.sum(roi * kernel)
    return output

# Define the input image and kernel
input_image = np.random.randn(48, 48)  # Assuming input size (48, 48)
kernel = np.random.randn(3, 3)  # Assuming kernel size (3, 3)
# Perform convolution with modified parameters
output = conv2d(input_image, kernel, padding=(1, 1))
# Now let's build a model equivalent to model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(48, 48, 1)))
model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(48, 48, 1)))

#2nd Convolution:

def conv2d(input_image, kernel, stride=(1, 1), padding=(0, 0)):
    # Get input image dimensions
    input_height, input_width = input_image.shape
    # Get kernel dimensions
    kernel_height, kernel_width = kernel.shape
    # Calculate output dimensions
    output_height = (input_height + 2 * padding[0] - kernel_height) // stride[0] + 1
    output_width = (input_width + 2 * padding[1] - kernel_width) // stride[1] + 1
    # Initialize output
    output = np.zeros((output_height, output_width))
    # Apply padding to input image
    padded_image = np.pad(input_image, ((padding[0], padding[0]), (padding[1], padding[1])), mode='constant')
    # Perform convolution
    for y in range(0, output_height):
        for x in range(0, output_width):
            # Apply stride
            y_start = y * stride[0]
            y_end = y_start + kernel_height
            x_start = x * stride[1]
            x_end = x_start + kernel_width
            # Extract region of interest from the padded image
            roi = padded_image[y_start:y_end, x_start:x_end]
            # Perform element-wise multiplication and sum
            output[y, x] = np.sum(roi * kernel)
    return output
# Define the input image and kernel
input_image = np.random.randn(48, 48)  # Assuming input size (48, 48)
kernel = np.random.randn(3, 3)  # Assuming kernel size (3, 3)
# Perform convolution
output = conv2d(input_image, kernel)
# Now let's build a model equivalent to model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))

model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

#3rd Convolution:

def conv2d(input_image, kernel, stride=(1, 1), padding=(0, 0)):
    # Get input image dimensions
    input_height, input_width = input_image.shape
    # Get kernel dimensions
    kernel_height, kernel_width = kernel.shape
    # Calculate output dimensions
    output_height = (input_height + 2 * padding[0] - kernel_height) // stride[0] + 1
    output_width = (input_width + 2 * padding[1] - kernel_width) // stride[1] + 1
    # Initialize output
    output = np.zeros((output_height, output_width))
    # Apply padding to input image
    padded_image = np.pad(input_image, ((padding[0], padding[0]), (padding[1], padding[1])), mode='constant')
    # Perform convolution
    for y in range(0, output_height):
        for x in range(0, output_width):
            # Apply stride
            y_start = y * stride[0]
            y_end = y_start + kernel_height
            x_start = x * stride[1]
            x_end = x_start + kernel_width
            # Extract region of interest from the padded image
            roi = padded_image[y_start:y_end, x_start:x_end]
            # Perform element-wise multiplication and sum
            output[y, x] = np.sum(roi * kernel)
    return output
# Define the input image and kernel
input_image = np.random.randn(48, 48)  # Assuming input size (48, 48)
kernel = np.random.randn(3, 3)  # Assuming kernel size (3, 3)
# Perform convolution
output = conv2d(input_image, kernel)
# Now let's build a model equivalent to model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))

#4th Convolution:

def conv2d(input_image, kernel, stride=(1, 1), padding=(0, 0)):
    # Get input image dimensions
    input_height, input_width = input_image.shape
    # Get kernel dimensions
    kernel_height, kernel_width = kernel.shape
    # Calculate output dimensions
    output_height = (input_height + 2 * padding[0] - kernel_height) // stride[0] + 1
    output_width = (input_width + 2 * padding[1] - kernel_width) // stride[1] + 1
    # Initialize output
    output = np.zeros((output_height, output_width))
    # Apply padding to input image
    padded_image = np.pad(input_image, ((padding[0], padding[0]), (padding[1], padding[1])), mode='constant')
    # Perform convolution
    for y in range(0, output_height):
        for x in range(0, output_width):
            # Apply stride
            y_start = y * stride[0]
            y_end = y_start + kernel_height
            x_start = x * stride[1]
            x_end = x_start + kernel_width
            # Extract region of interest from the padded image
            roi = padded_image[y_start:y_end, x_start:x_end]
            # Perform element-wise multiplication and sum
            output[y, x] = np.sum(roi * kernel)
    return output
# Define the input image and kernel
input_image = np.random.randn(48, 48)  # Assuming input size (48, 48)
kernel = np.random.randn(3, 3)  # Assuming kernel size (3, 3)
# Perform convolution
output = conv2d(input_image, kernel)
# Now let's build a model equivalent to model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))

model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

#5th Convolution:

def conv2d(input_image, kernel, stride=(1, 1), padding=(0, 0)):
    # Get input image dimensions
    input_height, input_width = input_image.shape
    # Get kernel dimensions
    kernel_height, kernel_width = kernel.shape
    # Calculate output dimensions
    output_height = (input_height + 2 * padding[0] - kernel_height) // stride[0] + 1
    output_width = (input_width + 2 * padding[1] - kernel_width) // stride[1] + 1
    # Initialize output
    output = np.zeros((output_height, output_width))
    # Apply padding to input image
    padded_image = np.pad(input_image, ((padding[0], padding[0]), (padding[1], padding[1])), mode='constant')
    # Perform convolution
    for y in range(0, output_height):
        for x in range(0, output_width):
            # Apply stride
            y_start = y * stride[0]
            y_end = y_start + kernel_height
            x_start = x * stride[1]
            x_end = x_start + kernel_width
            # Extract region of interest from the padded image
            roi = padded_image[y_start:y_end, x_start:x_end]
            # Perform element-wise multiplication and sum
            output[y, x] = np.sum(roi * kernel)
    return output
# Define the input image and kernel
input_image = np.random.randn(48, 48)  # Assuming input size (48, 48)
kernel = np.random.randn(3, 3)  # Assuming kernel size (3, 3)
# Perform convolution
output = conv2d(input_image, kernel)
# Now let's build a model equivalent to model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))

#6th Convolution:

def conv2d(input_image, kernel, stride=(1, 1), padding=(0, 0)):
    # Get input image dimensions
    input_height, input_width = input_image.shape
    # Get kernel dimensions
    kernel_height, kernel_width = kernel.shape
    # Calculate output dimensions
    output_height = (input_height + 2 * padding[0] - kernel_height) // stride[0] + 1
    output_width = (input_width + 2 * padding[1] - kernel_width) // stride[1] + 1
    # Initialize output
    output = np.zeros((output_height, output_width))
    # Apply padding to input image
    padded_image = np.pad(input_image, ((padding[0], padding[0]), (padding[1], padding[1])), mode='constant')
    # Perform convolution
    for y in range(0, output_height):
        for x in range(0, output_width):
            # Apply stride
            y_start = y * stride[0]
            y_end = y_start + kernel_height
            x_start = x * stride[1]
            x_end = x_start + kernel_width
            # Extract region of interest from the padded image
            roi = padded_image[y_start:y_end, x_start:x_end]
            # Perform element-wise multiplication and sum
            output[y, x] = np.sum(roi * kernel)
    return output
# Define the input image and kernel
input_image = np.random.randn(48, 48)  # Assuming input size (48, 48)
kernel = np.random.randn(3, 3)  # Assuming kernel size (3, 3)
# Perform convolution
output = conv2d(input_image, kernel)
# Now let's build a model equivalent to model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))

model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(4, activation='softmax'))
```

```python
#seeing predicted emotion and actual emotion of a randomly generated particular image of the testing dataset

label_dict = {0 : 'Angry', 1 : 'Happy', 2 : 'Neutral', 3 : 'Sad'}

figure = plt.figure(figsize=(20, 8))
for i, index in enumerate(np.random.choice(X_test.shape[0], size=24, replace=False)):
    ax = figure.add_subplot(4, 6, i + 1, xticks=[], yticks=[])
    ax.imshow(np.squeeze(X_test[index]))
    predict_index = label_dict[(y_pred[index])]
    true_index = label_dict[np.argmax(y_test,axis=1)[index]]
    
    ax.set_title("{} ({})".format((predict_index), 
                                  (true_index)),
                                  color=("green" if predict_index == true_index else "red"))
```
